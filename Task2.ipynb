```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\HP\Downloads\Finance_Analytics_Dataset.csv")
df
```

```python
df.head()
```

```python
sns.countplot(x="Category",data=df)
plt.show()
```

```python
sns.countplot(x="PaymentMode",data=df)
plt.show()
```

```python
sns.countplot(x="TransactionType",data=df)
plt.show()
```

```python

sns.barplot(data=df, x='TransactionType', y='Amount')
plt.title('Debit vs Credit Amount Comparison')
plt.xlabel('Transaction Type')
plt.ylabel('Amount')
```

```python
sns.barplot(data=df, x='CustomerSegment', y='Amount')
plt.title('Total spending by CustomerSegment')
plt.xlabel('Total Amount')
plt.show()
```

```python
payment_mode_counts = df['PaymentMode'].value_counts()
print(payment_mode_counts)
plt.pie(payment_mode_counts, labels=payment_mode_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Payment Mode Usage Distribution')
plt.show()
```

```python
sns.boxplot(data=df, x='Category', y='Amount')
plt.xticks(rotation=45)
plt.title('Transaction amount by Category')
plt.xlabel('Category')
plt.show()
```
